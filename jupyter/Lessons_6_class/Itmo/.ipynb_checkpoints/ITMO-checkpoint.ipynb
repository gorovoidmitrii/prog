Дезоксирибонуклеиновая кислота (ДНК) — один из двух типов нуклеиновых кислот, обеспечивающих хранение, передачу из поколения в поколение и реализацию генетической программы развития и функционирования живых организмов.

ДНК представляет собой пару полимерных молекул-цепей, каждое из звеньев которых является нуклеотидом одного из четырех видов: аденин (A), тимин (T), гуанин (G), или цитозин (C). На каждой из цепочек задано направление, причем для двух цепочек из одного ДНК направления всегда противоположны. Получается, что напротив первого нуклеотида одной цепочки находится последний нуклеотид другой, напротив второго — предпоследний и т.д.

Молекула ДНК устроена таким образом, что всегда соблюдается принцип комплементарности. Суть его состоит в том, что напротив аденина всегда находится тимин, и наоборот. Аналогичным образом гуанин соответствует цитозину. К примеру, цепочка CTGACA комплементарна цепочке TGTCAG.
В одной суперсекретной лаборатории у Миши целиком прочитали его ДНК, но отдали запись s только одной из двух цепочек. Помогите Мише найти вторую цепочку.
Входные данные
На вход подается строка s длиной не более 200 символов из букв «ATGC».

Выходные данные
Выведите последовательность, комплиментарную к заданной.

```python
Word = input()
listWord = []
for i in range(len(Word)):
    listWord.append(Word[i])

newWord='C'
for i in range(len(listWord)):
    if listWord[i] == 'C':
        newWord += 'G'
    elif listWord[i]== 'G':
        newWord += 'C'
    elif listWord[i]== 'A':
        newWord += 'T'
    else :
        newWord += 'A'
secondWord=newWord[1::]
secondWord=secondWord[::-1]
print(secondWord)
```
