```python
Два программиста ищут ошибки в большой программе. У каждого из них индивидуальный режим работы. Первый программист работает a часов, после чего отдыхает b часов. Второй программист работает c часов, после чего отдыхает d часов. Каждый из программистов находит по одной ошибке за каждый час работы.

Будем считать, что в настоящий момент оба программиста только что закончили отдыхать и одновременно начали работать. Через сколько часов все n ошибок будут найдены?

Входные данные
На вход программы поступают пять натуральных числа n, a, b, c и d, разделенные пробелами (1 ≤ n ≤ 2·1012, 1 ≤ a, b, c, d ≤ 10).

Выходные данные
Выведите одно число — число часов, после которых все n ошибок будут найдены.
```

```python
n, a, b, c, d = [int(elem) for elem in input().split()]
ACount = 0
BCount = 0
CCount = 0
DCount = 0
BugsCount = 0
BugsCount = (n//(a+b)*(c+d))*((a+b)*c+(c+d)*a)
FirstPrWorking = 1
SecondPrWorking = 1
HoursCount = n//(a+b)*(c+d)

NewN = n-BugsCount
while BugsCount < NewN:
    HoursCount += 1
    if FirstPrWorking == 1:
        if ACount < a: 
            BugsCount += 1
            ACount += 1
        else:
            ACount = 0
            BCount += 1
            FirstPrWorking = 0
    else:
        if BCount < b: 
            BCount += 1
        else:
            BCount = 0
            BugsCount += 1
            ACount += 1
            FirstPrWorking = 1
            
            
            
    if SecondPrWorking == 1:
        if CCount < c: 
            BugsCount += 1
            CCount += 1
        else:
            CCount = 0
            DCount += 1 
            SecondPrWorking = 0
    else:
        if DCount < d: 
            DCount += 1
        else:
            DCount = 0
            BugsCount += 1 
            CCount += 1
            SecondPrWorking = 1

print(HoursCount)    
       
```

```python
n = int(input())
lenghtList = [int(elem) for elem in input().split()]
m = int(input())
firstNewLenghtList = []
secondNewLenghtList = []
howManyCut = 0
howManyPieces = 0
for i in range (n):
    if lenghtList[i] % 10 == 0:
        firstNewLenghtList.append(lenghtList[i])
    else:
        secondNewLenghtList.append(lenghtList[i])

firstNewLenghtList.sort()


for j in range (len(firstNewLenghtList)):
    if howManyCut < m:
        if firstNewLenghtList[j] == 10:
            firstNewLenghtList[j] = 0
            howManyPieces += 1
        elif firstNewLenghtList[j] == 20:
            firstNewLenghtList[j] = 0
            howManyCut += 1
            howManyPieces += 2
        else:
            if firstNewLenghtList[j]>10:
                while firstNewLenghtList[j] >= 10:
                        if howManyCut < m:
                            firstNewLenghtList[j] -= 10
                            howManyCut += 1
                            howManyPieces += 1
                            if firstNewLenghtList[j] == 10:
                                howManyPieces += 1
                                firstNewLenghtList[j] = 0
                        else:
                            break
            
                
if howManyCut < m:
    for i in range (len(secondNewLenghtList)):
        if secondNewLenghtList[i]>10:
            while secondNewLenghtList[i] > 10:
                if howManyCut < m:
                    secondNewLenghtList[i] -= 10
                    howManyCut += 1
                    howManyPieces += 1
                    
                else:
                    break
                        
                    
print(howManyPieces)                               
                                
                        
                        
                
```
