```python
 
minutesInYear365 = 60*24*365
minutesInYear366 = 60*24*366
firstPartOfDate, secondPartOfDate=input().split(', ')
month, days = firstPartOfDate.split()
year, time = secondPartOfDate.split()
hours, minutes = time.split(':')
year = int(year)
hours = int(hours)
days = int(days)-1
minutes = int(minutes)

yearType = 0
if year % 400 == 0:
    yearType = 1
elif year % 4 == 0:
    if year % 100 != 0:
        yearType = 1
else:
    pass


def DayInYearCount(m):
    DayCount = 0
    if m == 'January':
        DayCount += 0
    if m == 'February':
        DayCount += 31
    

    if m == 'March':
        DayCount += 28+31
        
            
    if m == 'April':
        DayCount +=  31+28+31
        
            
    if m == 'May':
        DayCount += 30+31+28+31
        
            
    if m == 'June':
        DayCount += 28+31+30+31+31
        
            
    if m == 'July':
        DayCount += 28+31+30+31+31+30
        
            
    if m == 'August':
        DayCount += 28+31+30+30+31+31+31
        
            
    if m == 'September':
        DayCount += 28+31+30+30+31+31+31+31
        
    if m == 'October':
        DayCount += 28+31+30+30+30+31+31+31+31
        
    if m == 'November':
        DayCount += 28+31+30+30+30+31+31+31+31+31
        
    if m == 'December':
        DayCount += 28+31+30+30+30+30+31+31+31+31+31
    if yearType == 1:
        if m != 'January':
            if m != 'February':
                DayCount += 1 
    return DayCount


 
days+=DayInYearCount (month)
MinutesInDay = days*24*60
MinutesInHour= hours*60
minutes += MinutesInDay + MinutesInHour
if yearType == 0:
    print(minutes*100 / minutesInYear365)
else:
    print(minutes*100 / minutesInYear366)




#January 01, 2008 00:00

 #"January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December".
```
